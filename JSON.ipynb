```python
project_ls = []
category_ls = []
bid_ls = []
for index in range(0,len(data)):  
    URL = data[index]["URL"]
    organization = data[index]["organization"]
    
    if "utc_publish_date" in data[index]:
        publish_date = data[index]["utc_publish_date"]
    else:
        publish_date = None
        
    if "Bid Awarded Date" in data[index]:
        award_date = data[index]["Bid Awarded Date"]
    else:
        award_date = None
    
    if "Bid Type" in data[index]:
        Bid_Type = data[index]["Bid Type"]
    else:
        Bid_Type = None 
    
    if "Bid Number" in data[index]:
        Bid_Number = data[index]["Bid Number"]
    else:
        Bid_Number = None
    
    if "Bid Name" in data[index]:
        Bid_Name = data[index]["Bid Name"]
    else:
        Bid_Name = None
        
    if "Bid Classification" in data[index]:
        Bid_Classification = data[index]["Bid Classification"]
    else:
        Bid_Classification = None
    if Bid_Classification != "Construction":
        continue
    
    if "Bid Status" in data[index]:
        Bid_Status = data[index]["Bid Status"]
    else:
        Bid_Status = None 
    if Bid_Status != "Awarded":
        continue
    
#     if "Description" in data[index]:
#         Description = data[index]["Description"]
#     else:
#         Description = None
    
    if "Categories" in data[index]:
        Categories = data[index]["Categories"]
    else:
        Categories = None
    
    if "Awarded" in data[index]:
        Awarded = data[index]["Awarded"]
    else:
        Awarded = None
        
    if "Submitted" in data[index]:
        Submitted = data[index]["Submitted"]
    else:
        Submitted=None
        
    if "PlanTakers" in data[index]:
        PlanTakers = data[index]["PlanTakers"]
    else:
        PlanTakers = None
        
    project_ls.append((URL,organization,publish_date,award_date,Bid_Classification,Bid_Type,Bid_Number,Bid_Name,Bid_Status))
    category_ls.append((URL,Categories))
    bid_ls.append((URL,PlanTakers, Submitted, Awarded))
        
project_df = pd.DataFrame(project_ls, columns=["URL","organization","publish_date","award_date","Bid_Classification","Bid_Type","Bid_Number",
                                               "Bid_Name", "Bid_Status"])
categories_df = pd.DataFrame(category_ls, columns=["URL","Categories"])
bid_df = pd.DataFrame(bid_ls, columns=["URL","PlanTakers", "Submitted", "Awarded"])
```

```python
import re

award_df['Value'] = award_df['Value'].str.replace(',','')
award_df['Value'] = award_df['Value'].str.replace(' ','')


def split_it(value):
    return re.findall('\$+[0-9]*', value)

award_df['Value'] = award_df['Value'].apply(split_it)

def numberize(value):
    
    if len(value) != 0:
        value = [w.replace('$', '0') for w in value]
        value = list(map(int, value))
        value = sum(value)
        return value
    
    else:
        return None
    
award_df['Value'] = award_df['Value'].apply(numberize)
```
